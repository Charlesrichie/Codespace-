```python
#read the tsv file using pandas
# Importing the dataset
DATASET_COLUMNS  = ["sentiment", "ids", "date", "flag", "user", "tweet"]
DATASET_ENCODING = "ISO-8859-1"

dataset = pd.read_csv(io.BytesIO(uploaded['Sentiment140.tenPercent.sample.tweets.tsv']), sep = '\t', encoding=DATASET_ENCODING , names=DATASET_COLUMNS, header = None, engine = 'python')
# Dataset is now stored in a Pandas Dataframe

dataset.drop(index = 0, inplace = True)
dataset
```

```python
dataset.sentiment = dataset.sentiment.map({4:1,0:0})
```

```python
import seaborn as sns

sns.countplot(x='sentiment', data = dataset)
```

```python
# Remove duplicate tweets
init_count = len(dataset)
tweet_unique = dataset.tweet.drop_duplicates(keep=False)
records_with_same_tweet = dataset[True ^ dataset.tweet.isin(tweet_unique)]
records_with_same_tweet[['sentiment', 'user', 'tweet']]
dataset = dataset[True ^ dataset.tweet.isin(records_with_same_tweet.tweet)]
curr_count = len(dataset)
print(f"{init_count - curr_count} duplicate rows dropped!\nCurrent row count: {curr_count}\n\n")

dataset.sentiment.map({1:"Positive(1)",0:"Negative(0)"}).value_counts().plot(kind='bar', color=['green', 'red'])
plt.title("Data distribution", fontdict={"fontsize": 20})
plt.show()
```
